```python
import numpy as np
```

```python
data = [99,86,87,88,111,86,103,87,94,78,77,85,86] 
```

```python
data_mean = np.mean(data)
print(data_mean)
```

```python
data_median = np.median(data)
print(data_median)
```

```python
from scipy import stats
```

```python
data_mode = stats.mode(data)
print(data_mode)
```

```python
data_std = np.std(data)
print(data_std)
```

```python
data_var = np.var(data)
print(data_var)
```
