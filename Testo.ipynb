```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
df = pd.read_csv("dataset/Salary_dataset.csv")
df
```

```python
df = df.drop('Unnamed: 0', axis=1)
df
```

```python
plt.title("Connection between Years Experience and Salary")
plt.scatter(df['YearsExperience'], df['Salary'])
plt.xlabel("Years Experience")
plt.ylabel("Salary")
plt.show()
```

```python
X = df[["YearsExperience"]]
y = df["Salary"]
```

```python
data = LinearRegression()
```

```python
data.fit(X, y)
```

```python
print("Coefficient: ", data.coef_)
print("Intercept: ", data.intercept_)
```

```python
print("Prediction:", data.predict([[2]]))
```

```python
plt.title("Linear Regression Model")
plt.plot(X, data.predict(X))
plt.scatter(X, y)
plt.show
```
