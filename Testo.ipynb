```python
import numpy as np
import matplotlib.pyplot as plt

# Data non-linear (kurva)
np.random.seed(0)
X = np.linspace(0, 5, 20).reshape(-1, 1)
y = X.ravel()**2 + np.random.randn(20) * 2
```

```python
from sklearn.linear_model import LinearRegression

model_underfit = LinearRegression()
model_underfit.fit(X, y)

X_test = np.linspace(0, 5, 100).reshape(-1, 1)
y_pred_underfit = model_underfit.predict(X_test)
```

```python
from sklearn.preprocessing import PolynomialFeatures
from sklearn.pipeline import make_pipeline

model_good = make_pipeline(
    PolynomialFeatures(degree=2),
    LinearRegression()
)

model_good.fit(X, y)
y_pred_good = model_good.predict(X_test)
```

```python
plt.scatter(X, y, color='black', label='Data')

plt.plot(X_test, y_pred_underfit, label='Underfitting (Linear)', linestyle='--')
plt.plot(X_test, y_pred_good, label='Good Fit (Polynomial Degree 2)')

plt.legend()
plt.title("Contoh Underfitting")
plt.show()
```
