```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
X = np.array([
    [50, 2, 10],
    [60, 3, 8],
    [80, 3, 5],
    [100, 4, 2],
    [120, 4, 1]
])

y = np.array([300, 360, 420, 500, 550])
```

```python
model = LinearRegression()
model.fit(X, y)
```

```python
print("Intercept: ", model.intercept_)
print("Coefficient: ", model.coef_)
```

```python
prediction = model.predict([[90, 3, 4]])
print(prediction)
```

```python
r2 = model.score(X, y)
print("R^2: ", r2)
```

```python
plt.title("Linear Regression")
plt.scatter(X[:, 0], y)
plt.scatter(X[:, 1], y)
plt.scatter(X[:, 2], y)
plt.plot(X, model.predict(X))
plt.show()
```
